### **Edge detection using Canny Algorithm**

- Canny edge detection is based on:

 - gradient of the image i.e. **difference between two adjacent pixels**

 - **hysteresis filtering:** It selects the lines using those pixels which are different from adjacent ones.

- For the Canny edge detection algorithm, we need to provide **3 arguments** in the **cv2.Canny()** function i.e. input **image array, the minimum value of a pixel, the maximum value of pixel.**

**Method 1**

```python
plt.figure(figsize=(15,15))

edges = cv2.Canny(img_color, 100, 200)
plt.imshow(edges, cmap = 'gray')

plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

**Method 2**

```python
from google.colab.patches import cv2_imshow

# Canny Edge Detection
edges = cv2.Canny(image=img_color, threshold1=100, threshold2=200)

# Display Canny Edge Detection Image
cv2_imshow(edges)
cv2.waitKey(0)
```

#### **2) Sobel**

**Syntax**

Sobel(src, ddepth, dx, dy)

```python
from google.colab.patches import cv2_imshow

# Sobel Edge Detection
sobelx = cv2.Sobel(src=img_color, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5) # Sobel Edge Detection on the X axis
sobely = cv2.Sobel(src=img_color, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5) # Sobel Edge Detection on the Y axis
sobelxy = cv2.Sobel(src=img_color, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5) # Combined X and Y Sobel Edge Detection

# Display Sobel Edge Detection Images
cv2_imshow(sobelx)
cv2.waitKey(0)

cv2_imshow(sobely)
cv2.waitKey(0)

cv2_imshow(sobelxy)
cv2.waitKey(0)
```
