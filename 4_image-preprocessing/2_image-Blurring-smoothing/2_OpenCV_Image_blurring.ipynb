```python
# importing libraries
import cv2
import numpy as np

from google.colab.patches import cv2_imshow
```

```python
path = '/content/drive/MyDrive/OpenCV-Images/fruit.JPG'

image = cv2.imread(path)
cv2_imshow(image)
```

```python
image = cv2.imread(path)
  
cv2_imshow(image)  # Original Image
cv2.waitKey(0)

# Averaging
Averaging = cv2.blur(image, (10, 10))
cv2_imshow(Averaging)   # Averaging
cv2.waitKey(0)
  
# Gaussian Blur
Gaussian = cv2.GaussianBlur(image, (7, 7), 0)
cv2_imshow(Gaussian)   # Gaussian Blurring
cv2.waitKey(0)
  
# Median Blur
median = cv2.medianBlur(image, 5)
cv2_imshow(median)  # Median Blurring
cv2.waitKey(0)
  
  
# Bilateral Blur
bilateral = cv2.bilateralFilter(image, 9, 75, 75)
cv2_imshow(bilateral)  # Bilateral Blurring
cv2.waitKey(0)
cv2.destroyAllWindows()
```
